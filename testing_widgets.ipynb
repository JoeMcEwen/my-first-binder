```python
import ipywidgets as widgets
from ipywidgets import interact, IntSlider

import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt

from numpy import pi, cos, sin, exp
```

```python
def f(x):
    return 2*x
```

```python
z=interact(f,x=10)
```

```python
def plot_harmonic(n):
    
    nx=100
    L=1
    
    x=np.linspace(0,L,nx)
    
    kn=2*pi*n/L
    
    harmonic=sin(kn*x)
    
    fig=plt.figure()
    ax=fig.add_subplot(111)
    
    ax.plot(x,harmonic)
    ax.grid()
    
    
```

```python
plot_harmonic(4)
```

```python
interact(plot_harmonic,n=IntSlider(min=1,max=100,n=1));
```
